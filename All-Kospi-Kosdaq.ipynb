## 재무제표 갖고 오기

```python
def get_finacial_statements_url(code, ifrs_type=4, unit='Y'):
    url_tmpl = 'http://companyinfo.stock.naver.com/v1/company/ajax/cF1001.aspx?cmp_cd=%s&fin_typ=%s&freq_typ=%s'
    url = url_tmpl % (code, ifrs_type, unit) # 삼성전자, 4(IFRS 연결), Y:년 단위
    print(url)
    dfs = pd.read_html(url, encoding='utf8')    
    # dfs
    df = dfs[0]
    # print(df)
    df = df.set_index('주요재무정보')
    return df
```

```python
# 리스트로 전환
def get_financial_statements_infos(df):
    cols = []
    for year in range(2012, 2020):
        cols.append(str(year) + '/12')    
    df.columns = cols

    indexes = []
    for i in range(0, df.index.size):
        indexes.append(df.index[i][0])
        
    new_index_dict = dict(zip(df.index, indexes))
    df.rename(new_index_dict, axis='index', inplace=True)  
    
    df.fillna(0, inplace=True)
    
    dft = df.T

    # 첫번째 컬럼 이름을 data로 변경
    dft.rename(columns={'주요재무정보':'date'}, inplace=True)

    # 인덱스를 날짜형식(datetime)으로 변환
    dft.index = pd.to_datetime(dft.index)
    
    return dft
```

```python
# 증자 여부: 증자기업 나쁜 기업. 즉, '증자여부'가 0이어야지 됨.
def check_increase_capital(df):
#     df['자본금'] = df['자본금'].astype(float).fillna(0.0)
    df['증자여부'] = 0.0
    for index in range(1, len(df.index)):
        df.loc[index:index+1,'증자여부'] = df.iloc[index-1]['자본금'] - df.iloc[index]['자본금']
```
