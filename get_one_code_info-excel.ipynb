```python
import pandas as pd
```

```python
CODE = '000220'
COMPANY_NM = '기신정기'
```

```python
# iframe 네이버 재무제표: http://finance.naver.com/item/coinfo.nhn?code=005930
url_tmpl = 'http://companyinfo.stock.naver.com/v1/company/ajax/cF1001.aspx?cmp_cd=%s&fin_typ=%s&freq_typ=%s'
url = url_tmpl % (CODE, '4', 'Y') # 삼성전자, 4(IFRS 연결), Y:년 단위, Q: 분기
print(url)

dfs = pd.read_html(url, encoding='utf8')
# dfs
df = dfs[0]
# print(df)
# df = df.set_index('주요재무정보')
df.head(5) # 10개 항목만 표시(실제 32개 항목)
```

# 컬럼 변환

```python
# multi columns 제거
cols = list(df.columns.droplevel(0))
cols
```

```python
# 재무제표 년/월 얻어오기
import re
from datetime import datetime

s = cols[0]

def get_date_str(s):
    date_str = ''
    r = re.search("\d{4}/\d{2}", s)
#     r = re.search("\d{4}", s)
    if r:
        date_str = r.group()

    return date_str
```

```python
cols = [get_date_str(col) for col in cols]
cols
```

```python
df = df.set_index('주요재무정보')
df.head()
```

```python
# 컬럼이름 변경
df.columns = cols[0:-1]
df.head()
```

## Index 변환

```python
indexes = []

for i in range(0, df.index.size):
    indexes.append(df.index[i][0])
    
indexes
```

```python
# new_indexes = {v: k for v, k in enumerate(indexes)}

new_index_dict = dict(zip(df.index, indexes))
print(new_index_dict)

df.rename(new_index_dict, axis='index', inplace=True)
df.head(5)
```

# NaN을 0.0 숫자로 바꾸기

```python
df.fillna(0, inplace=True)
df.head()
```

# excel 저장

```python
def get_column_values(df, years):
    found = []
    for col in df.columns.values:
        if any(year in col for year in years):
            found.append(col)
    return found
```

```python
columns = get_column_values(df, ('2016', '2017'))

df[columns].head()
```

```python
# 항목별로 다른 sheet에 저장하기
def save_fincial_statement(file_name, sheet_names, inquiry_years, code, corp_nm):
    # Create a Pandas Excel writer using XlsxWriter as the engine.
    with pd.ExcelWriter(file_name, engine='xlsxwriter') as writer:
        for idx, val in enumerate(sheet_names):
            data = df[idx:idx+1][inquiry_years]
            data.insert(0, 'corp_nm', corp_nm)
            data.index.rename("", inplace=True)
            data.index = [code]
            data.to_excel(writer, index=True, sheet_name=sheet_names[idx])
        writer.save()
```

```python
# sample.xlsx에 2016/12, 2017/12 데이터 저장하기
years = get_column_values(df, ('2016','2017'))
print(years)
save_fincial_statement('sample.xlsx', df.index.values, years, CODE, COMPANY_NM)
```

### 엑셀 마지막 row에 추가

```python
from openpyxl import load_workbook
```

```python
book = load_workbook('sample.xlsx')
writer = pd.ExcelWriter('sample.xlsx', engine='openpyxl')
writer.book = book
for sheet_name in df.index.values:
    writer.book[sheet_name].append([CODE, COMPANY_NM, df.loc[sheet_name]['2016/12'], df.loc[sheet_name]['2017/12']])
writer.save()
```

```python
def append_financial_statement(df, file_name, *datas):
    book = load_workbook(file_name)
    writer = pd.ExcelWriter(file_name, engine='openpyxl')
    writer.book = book
    for sheet_name in df.index.values:
        writer.book[sheet_name].append(datas)
    writer.save()
```

```python
append_financial_statement(df, 'sample.xlsx', CODE, COMPANY_NM, df.loc['매출액']['2016/12'], df.loc['매출액']['2016/12'])
```
