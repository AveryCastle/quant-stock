```python
import pandas as pd
```

```python
CODE = '005930'
```

```python
url_tmpl = 'http://companyinfo.stock.naver.com/v1/company/ajax/cF1001.aspx?cmp_cd=%s&fin_typ=%s&freq_typ=%s'
url = url_tmpl % (CODE, '4', 'Y') # 삼성전자, 4(IFRS 연결), Y:년 단위
print(url)

dfs = pd.read_html(url, encoding='utf8')
# dfs
df = dfs[0]
# print(df)
# df = df.set_index('주요재무정보')
df.head(10) # 10개 항목만 표시(실제 32개 항목)
```

# 컬럼 변환

```python
# multi columns 제거
cols = list(df.columns.droplevel(0))
cols
```

```python
# 재무제표 년/월 얻어오기
import re
from datetime import datetime

s = cols[0]

def get_date_str(s):
    date_str = ''
    r = re.search("\d{4}/\d{2}", s)
    if r:
        date_str = r.group()

    return date_str
```

```python
cols = [get_date_str(col) for col in cols]
cols
```

```python
df = df.set_index('주요재무정보')
df.head()
```

```python
# 컬럼이름 변경
df.columns = cols[0:-1]
df.head()
```

## Index 변환

```python
indexes = []

for i in range(0, df.index.size):
    indexes.append(df.index[i][0])
    
indexes
```

```python
# new_indexes = {v: k for v, k in enumerate(indexes)}

new_index_dict = dict(zip(df.index, indexes))
print(new_index_dict)

df.rename(new_index_dict, axis='index', inplace=True)
df.head(5)
```

# NaN을 0.0 숫자로 바꾸기

```python
df.fillna(0, inplace=True)
df.head()
```

# excel 저장

```python
# 항목별로 다른 sheet에 저장하기
def save_fincial_statement(file_name, sheet_names, inquiry_years, code):
    # Create a Pandas Excel writer using XlsxWriter as the engine.
    with pd.ExcelWriter(file_name, engine='xlsxwriter') as writer:
        for idx, val in enumerate(sheet_names):
            data = df[idx:idx+1][inquiry_years]
            data.index.rename("", inplace=True)
            data.index = [code]
            data.to_excel(writer, index=True, sheet_name=sheet_names[idx])
        writer.save()
```

```python
# sample.xlsx에 2016/12, 2017/12 데이터 저장하기
save_fincial_statement('sample.xlsx', df.index.values, ['2016/12','2017/12'], CODE)
```
