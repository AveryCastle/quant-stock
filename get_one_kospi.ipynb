```python
import pandas as pd
```

```python
url_tmpl = 'http://companyinfo.stock.naver.com/v1/company/ajax/cF1001.aspx?cmp_cd=%s&fin_typ=%s&freq_typ=%s'
url = url_tmpl % ('005930', '4', 'Y') # 삼성전자, 4(IFRS 연결), Y:년 단위

dfs = pd.read_html(url, encoding='utf8')
# dfs
df = dfs[0]
# print(df)
df = df.set_index('주요재무정보')
df.head(10) # 10개 항목만 표시(실제 32개 항목)
```

# 컬럼 변환

```python
# 리스트로 전환
cols = []

for year in range(2012, 2020):
    cols.append(str(year) + '/12')
    
cols
```

```python
df.columns = cols
df.head()
```

# Index  변환

```python
indexes = []

for i in range(0, df.index.size):
    indexes.append(df.index[i][0])
    
indexes
```

```python
df.head(5)
```

```python
# new_indexes = {v: k for v, k in enumerate(indexes)}

new_index_dict = dict(zip(df.index, indexes))
print(new_index_dict)

df.rename(new_index_dict, axis='index', inplace=True)
df.head(5)
```

# NaN을 0.0 숫자로 바꾸기

```python
df.fillna(0, inplace=True)
df.head()
```

# column & row change

```python
dft = df.T

# 첫번째 컬럼 이름을 data로 변경
dft.rename(columns={'주요재무정보':'date'}, inplace=True)

# 인덱스를 날짜형식(datetime)으로 변환
dft.index = pd.to_datetime(dft.index)

dft[['매출액', '당기순이익', '부채총계', 'PER(배)' , 'PBR(배)', '현금배당수익률', '자본금']]
```

```python
# 증자 여부: 증자기업 나쁜 기업. 즉, '증자여부'가 0이어야지 됨.
def check_increase_capital(df):
    df['증자여부'] = 0.0
    for index in range(1, len(dft.index)):
        df.loc[index:index+1,'증자여부'] = df.iloc[index-1]['자본금'] - dft.iloc[index]['자본금']
```

```python
check_increase_capital(dft)
```
